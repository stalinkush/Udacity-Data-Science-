```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#A fair coin flip produces heads

#Stimulating taking 1 million samples
samples = np.random.binomial(n=1, p=0.5, size=int(1e6))
sns.set()
plt.hist(samples, normed=True)
plt.title('Probability of fair coin flip produces heads')
(samples == 1).mean()
```

```python
#Five fair coin flips produce exactly one head

#Stimulating taking 1 million samples
samples_1 = np.random.binomial(n=5, p=0.5, size=int(1e6))
plt.hist(samples_1, bins=100)
(samples_1 == 1).mean()
```

```python
#Ten fair coin flips produce exactly four heads

#Stimulating taking 1 million samples
samples_2 = np.random.binomial(n=10, p=0.5, size=int(1e6))
plt.hist(samples_2)
(samples_2 == 4).mean()
```

```python
#Five biased coin flips with P(H) = 0.8 produce exactly five heads

#Stimulating taking 1 million samples
samples_3 = np.random.binomial(n=5, p=0.8, size=int(1e6))
plt.hist(samples_3)
(samples_3 == 5).mean()
```

```python
#Ten biased coin flips with P(H) = 0.15 produce exactly three heads

#Stimulating taking 1 million samples
samples_4 = np.random.binomial(n=10, p=0.15, size=int(1e6))
plt.hist(samples_4)
(samples_4 == 3).mean()
```
