### The Impact of Large Sample Sizes

When we increase our sample size, even the smallest of differences may seem significant.  

To illustrate this point, work through this notebook and the quiz questions that follow below.

Start by reading in the libraries and data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
np.random.seed(42)
sns.set()

full_data = pd.read_csv('coffee_dataset.csv')
full_data.head()
```

`1.` In this case, imagine we are interested in testing if the mean height of all individuals in `full_data` is equal to 67.60 inches.  First, use **quiz 1** below to identify the null and alternative hypotheses for these cases. 

$$H_0: \mu = 67.60$$

$$H_1: \mu \neq 67.60$$

`2.` What is the population mean height?  What is the standard deviation of the population heights?  Create a sample set of data using the code below.  What is the sample mean height?  Simulate the sampling distribution for the mean of five values to see the shape and plot a histogram.  What is the standard deviation of the sampling distribution of the mean of five draws? Use **quiz 2** below to assure your answers are correct.  

```python
pop_height = full_data['height'].mean()
pop_height_std = np.std(full_data['height'])

print(' The Mean Population Height is {:.4f}\n The Std of Population is {:.4f}'.format(pop_height, pop_height_std))
```

```python
sample5 = full_data.sample(5)
obs_mean = sample5['height'].mean()
print('The Mean Height in Sample of size 5 is {:.4f}'.format(obs_mean))  
```

```python
#Stimulating Sampling Distribution of sample size 5
bootstrap_samples_5 = []

for i in range(10000):
    sample = sample5.sample(5, replace=True)
    sample_mean = sample['height'].mean()
    bootstrap_samples_5.append(sample_mean)

bootstrap_samples_5 = np.array(bootstrap_samples_5)
```

```python
plt.figure(figsize=(10, 8))
plt.hist(bootstrap_samples_5, color='#001f3f', alpha=0.6, bins=15)
plt.xlabel('Population Height', fontsize=15)
plt.title('Bootstrap Sampling of Sample size 5', fontsize=20)
plt.axvline(bootstrap_samples_5.mean(), linestyle='--', color='k', label='Mean = {:.4f}'.format(bootstrap_samples_5.mean()))
plt.legend(loc = 'upper left', frameon=True, fontsize=15, framealpha=0.6, facecolor='#0074D9');
```

```python
sample5_std = np.std(bootstrap_samples_5)
print( 'The Std of Sampling Distribution of sample size 5 is {:.4f}'.format(sample5_std)) 
```

`3.` Using the null and alternative set up in question 1 and the results of your sampling distribution in question 2, simulate the mean values you would expect from the null hypothesis.  Use these simulated values to determine a p-value to make a decision about your null and alternative hypotheses.  Check your solution using **quiz 3** and **quiz 4** below.

**Hint:** Use the numpy documentation [here](https://docs.scipy.org/doc/numpy-1.13.0/reference/generated/numpy.random.normal.html) to assist with your solution.

```python
#Distribution of Null Hypothesis
null_mean = 67.60
null_vals = np.random.normal(null_mean, sample5_std, size=len(bootstrap_samples_5))
```

```python
# for a two sided hypothesis, we want to look at anything 
# more extreme from the null in both directions
upper_extreme = (null_vals > obs_mean).mean()
lower_extreme = (null_vals < null_mean - (obs_mean - null_mean)).mean()

p = lower_extreme + upper_extreme

lower = null_mean - (obs_mean - null_mean)
upper = obs_mean

plt.figure(figsize=(10, 8))
plt.hist(null_vals, bins=50, color='#393f4d')
plt.xlabel('Population Heights', fontsize=15)
plt.title('(Null Hypothesis) Sampling Distribution\n of Null values', fontsize=20)
plt.axvline(lower, color='r', linestyle='--')
plt.axvline(upper, color='r', linestyle='--')
plt.text(s='P-value = {:.4f}'.format(p), x=63.5, y=625, bbox=dict(color='#fa625f', alpha=0.6), fontsize=15); 
```

```python
sample1000_std = np.std(bootstrap_samples_1000)
print('The Std of Sampling Distribution of sample size 1000 is {:.4f}'.format(sample1000_std)) 
```

```python
plt.figure(figsize=(10, 8))
plt.hist(bootstrap_samples_1000, color='#001f3f', alpha=0.6, bins=15)
plt.xlabel('Population Height', fontsize=15)
plt.title('Bootstrap Sampling of Sample size 1000', fontsize=20)
plt.axvline(bootstrap_samples_1000.mean(), linestyle='--', color='k', label='Mean = {:.4f}'.format(bootstrap_samples_1000.mean()))
plt.legend(loc = 'upper left', frameon=True, fontsize=15, framealpha=0.6, facecolor='#0074D9');
```

```python
#Null values
null_vals1 = np.random.normal(null_mean, sample1000_std, size=len(bootstrap_samples_1000)) 
```

```python
# for a two sided hypothesis, we want to look at anything 
# more extreme from the null in both directions
up_ext = (null_vals1 > obs_mean).mean()
low_ext = (null_vals1 < null_mean - (obs_mean - null_mean)).mean()

p1 = up_ext + low_ext

lower1 = null_mean - (obs_mean - null_mean)
upper = obs_mean

plt.figure(figsize=(10, 8))
plt.hist(null_vals1, bins=50, color='#393f4d')
plt.xlabel('Population Heights', fontsize=15)
plt.title('(Null Hypothesis) Sampling Distribution\n of Null values', fontsize=20)
plt.axvline(lower, color='r', linestyle='--')
plt.axvline(upper, color='r', linestyle='--')
plt.text(s='P-value = {:.4f}'.format(p1), x=67.25, y=600, bbox=dict(color='#fa625f', alpha=1), fontsize=15);   
```
